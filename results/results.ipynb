```python
import matplotlib.pyplot as plt

# convert the previous lines into a function
def plotMetrics(metricsQueries:list, metrics, label):
    plt.figure(figsize=(8,5))
    # print the metrics of the first query
    plt.bar(metrics, metricsQueries[0], label=label, width=0.4)
    for i in range(len(metricsQueries)-1):
        plt.bar(metrics, metricsQueries[i+1], label=label + ' optimized V {}'.format(i+1), width=0.4, bottom=metricsQueries[i])
    plt.legend()
    plt.title('Comparison of ' + label + ' metrics')
    plt.xlabel('Metrics')
    plt.ylabel('Values')
    plt.show()
    
```
