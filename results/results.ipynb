```python
import matplotlib.pyplot as plt

# convert the previous lines into a function
def plotMetrics(metricsQueries:list, metrics, label):
    plt.figure(figsize=(8,5))
    plt.bar(metrics, metricsQueries[0], label=label, width=0.4)
    plt.bar(metrics, metricsQueries[1], label=label + ' (Optimized)', width=0.4, bottom=metricsQueries[0])
    plt.bar(metrics, metricsQueries[2], label=label + ' (With Indexing)', width=0.4, bottom=metricsQueries[1])
    plt.bar(metrics, metricsQueries[3], label=label + ' (With Caching)', width=0.4, bottom=metricsQueries[2])
    plt.bar(metrics, metricsQueries[4], label=label + ' (With Schema Optimization)', width=0.4, bottom=metricsQueries[3])
    plt.bar(metrics, metricsQueries[5], label=label + ' (With Full Optimization)', width=0.4, bottom=metricsQueries[4])
    

    plt.legend()
    plt.title('Comparison of ' + label + ' metrics')
    plt.xlabel('Metrics')
    plt.ylabel('Values')
    plt.show()

metricsSQL = ['planning_time', 'execution_time']
```
